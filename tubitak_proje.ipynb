```python
def calculate_carbon_footprint(food_name, amount_kg):
    carbon_data = {
        "dana eti": 27.0,
        "tavuk eti": 6.9,
        "mercimek": 0.9,
        "elma": 0.4,
        "süt": 1.9
    }

    factor = carbon_data.get(food_name.lower(), None)

    if factor is None:
        return None
    else:
        total_carbon = factor * amount_kg
        return round(total_carbon, 2)


print("--- Karbon Ayak İzi Hesaplayıcı ---")
print("Mevcut gıdalar: Dana eti, Tavuk eti, Mercimek, Elma, Süt")

user_food = input("Hangi gıdayı tükettiniz?: ")

try:
    # Miktarı sayıya çevir, harf olursa yakalar
    user_amount = float(input(f"Kaç kg {user_food} tükettiniz?: "))

    result = calculate_carbon_footprint(user_food, user_amount)

    if result is not None:
        print(f"\n Sonuç: {user_amount} kg {user_food} tüketimi yaklaşık {result} kg CO2 salınımına neden olur.")
    else:
        print(f"\n Üzgünüz, '{user_food}' adlı gıda henüz listemizde bulunmuyor.")

except ValueError:
    print("\n Hata: Lütfen miktar kısmına sadece sayı giriniz (Örn: 0.5 veya 2).")
```
